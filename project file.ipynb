```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
rgb_path = r"C:\Users\Ankush\Desktop\Btech\Capstone\extracted\images"
depth_path = r"C:\Users\Ankush\Desktop\Btech\Capstone\extracted\depths"
label_path = r"C:\Users\Ankush\Desktop\Btech\Capstone\extracted\labels"
```

```python
print("RGB folder exists:", os.path.exists(rgb_path))
print("Depth folder exists:", os.path.exists(depth_path))
print("Label folder exists:", os.path.exists(label_path))
```

```python
files = os.listdir(rgb_path)
sample_name = files[0]

print("\nLoading file:", sample_name)

rgb = cv2.imread(os.path.join(rgb_path, sample_name))
rgb = cv2.cvtColor(rgb, cv2.COLOR_BGR2RGB)

depth = cv2.imread(os.path.join(depth_path, sample_name), cv2.IMREAD_GRAYSCALE)

label = cv2.imread(os.path.join(label_path, sample_name), cv2.IMREAD_GRAYSCALE)
```

```python
print("\nImage Shapes:")
print("RGB shape:", rgb.shape)
print("Depth shape:", depth.shape)
print("Label shape:", label.shape)
```

```python
print("\nDepth Range:")
print("Min:", depth.min())
print("Max:", depth.max())
```

```python
print("\nUnique Label Classes:")
print(np.unique(label))
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(rgb)
plt.title("RGB Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(depth, cmap="gray")
plt.title("Depth Map")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(label, cmap="jet")
plt.title("Segmentation Label")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
print("Unique label values:", np.unique(label))
```
